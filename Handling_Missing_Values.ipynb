```python
# Handling missing values
# METHOD 1: DELETTE ROWS AND COLUMNS
updated_df= df.dropna(axis=0)
print(df.shape)
print(updated_df.shape)
updated_df.info()
# And if you don't define any axis That means 👉
# it removes rows that contain missing (NaN) values.
```

```python
# METHOD 2 Replacing with mean median and mode
# We can use mode for gender and surname
# we can use median for credit score
# We can use mean for and balance
df=pd.read_excel("Churn_Modelling.csv.xlsx")
updated_df= df
print(updated_df["Surname"].mode()) # we can use this but better practice is to drop it

df.info()
```

```python
updated_df.dropna(subset=["Surname"], inplace=True)
```

```python

mode = updated_df["Exited"].mode()[0]
print(mode)
print('\n')

updated_df.fillna({"Exited": mode},inplace=True)

updated_df.info()
```

```python
median = updated_df["CreditScore"].median
print(median ,"\n")
updated_df.fillna({"CreditScore": median}, inplace=True)
updated_df.info()
```

```python
mode =updated_df["Gender"].mode
print(mode,"\n")
updated_df.fillna({"Gender": mode}, inplace=True)
updated_df.info()
```

```python
mean =updated_df["Age"].mean
updated_df.fillna({"Age":mean}, inplace=True)
updated_df.info()
```

```python
mean2 =updated_df["Balance"].mean
updated_df.fillna({"Balance":mean2}, inplace=True)
updated_df.info()
```

```python
mode = updated_df["HasCrCard"].mode()[0]
print(mode)
updated_df.fillna({"HasCrCard": mode}, inplace=True)
updated_df.info()
```

```python
# final check
print(updated_df.isnull().sum())
```

Forward Fill (ffill)
👉 Concept:

Replace missing value with the previous valid value.

```python
df=pd.read_excel("Churn_Modelling.csv.xlsx")
print(df.isnull().sum())
df.fillna(method="ffill", inplace=True)
print(df.isnull().sum())
```

 Backward Fill (bfill)
👉 Concept:

Replace missing value with the next valid value.

```python
df=pd.read_excel("Churn_Modelling.csv.xlsx")
print(df.isnull().sum())
df.fillna(method="bfill", inplace=True)
print(df.isnull().sum())
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/Anannya-Vyas/datasets/main/Churn_Modelling.xlsx"

ndf = pd.read_excel(url)
print(ndf.isnull().sum())



# backward fill only for Surname column
ndf["Surname"] = ndf["Surname"].bfill()
print(ndf.isnull().sum())
```

```python
ndf.head()
```

```python
# forward fill only for Age column
ndf["Age"] = ndf["Age"].ffill()
print(ndf.isnull().sum())
```
